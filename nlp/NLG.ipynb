## 3. NLG系统结构:

**NLG 系统结构大致分为三步骤：**

<img src="TP_github/NLG_architecture.png" width="500" height="500">

1. Document planning：如何从大量信息中选择重要信息以及如何确定条理清晰的文章结构。
    * Theoretical approach
    * Pragmatic approach
    * Statistical approach
2. Microplanning：使用哪些单词，句子表达。

3. Realisation：具体的语法细节，例如：单复数。

### 3.1 Document planning 如何选择内容以及如何确定结构？

内容选择是自然语言生成中最重要的部分，并且内容的选择与所涉及的领域有紧密关联。内容的选择有三种方法：Theoretical approach，Pragmatic approach，Statistical approach。结构的确定可以使用一些连接词，例如：concession, contrast, elaboration, example, reason, sequence等。

<img src="TP_github/example_NLG.png" width="500" height="500">
<img src="TP_github/example1_NLG.png" width="500" height="500">

描述现象，对于特殊存在危险的运动轨迹解释其潜在风险并给出建议。通常，将其分为几个部分，对每一部分描述其最重要的安全问题，然后描述略微次要的问题，最后将其连接到一起通过一些连词的使用。更高级的自然语言生成可以根据阅读者的特点喜好从而生成不同风格的语言，例如：一些阅读者更喜欢积极的信息，一些阅读者更喜欢直接简洁的表达，然而，一些人喜欢间接的表达等。

### 3.2.1 词汇的选择：Frequency，Formality，Technical terms，Convention，expectations

例如：Statistics-Based Lexical Choice for NLG from Quantitative Information

例如下面的每组数据由三个属性组成。

<img src="TP_github/weather_forecast.png" width="500" height="500">

将每一个属性（例如：wind speed）由一系列由权重所组成的向量表示。例如：wind speed = 9 = [0.1, 0.9, 0, 0, 0]

<img src="TP_github/wind_speed.png" width="500" height="500">

将向量压缩到一起 {ws = 9,dir = 2} → [0.1,0.9,0,0,0,0.97,0.03,0,0,0]

使用矩阵形式表示所有数据。

<img src="TP_github/example_matrix.png" width="500" height="500">

以风速为例，对于不同的wind speed可以由不同的单词表示：breeze, strong, storm等。将每个单词通过相同长度的向量表示，通过计算单词向量与数据向量的cosin值获得相似度，最后，将未知数据的向量与每一个单词向量比较，使用相似度最大的单词表示此数据。

计算单词向量公式如下：
$$
\text{V}_{i} = min \sqrt{\sum_d \left(\text{V}_i \text{V}_d - appear(\text{V}_i, \text{V}_d)\right)^2}
$$

其中appear代表此单词是否在某一文本中出现，出现为1，否则为0，通过矩阵形式表示，每一列 $e_i$ 代表每一个单词在所有文本中出现情况：

<img src="TP_github/appearance.png" width="500" height="400">

使用矩阵形式表示上述等式：

$$
\text{V}_{i} = min \sqrt{\left(\text{K} \text{V}_i - \text{e}_i \right)^2}
$$

根据最小二乘法得：

$$
\text{V}_i = (\text{K}^T \text{K})^{-1} \text{K}^T e_i
$$
