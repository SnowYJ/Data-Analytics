## 2. 逻辑回归：

### 1. odds与logit函数：

**一个事件的odds为该事件发生的几率p与该事件未发生的几率1-p的比值。**

$$odds = \frac{p}{1-p}$$

几率的取值范围为0到正无穷。对odds取log，图像如下：
   
<img src="ml_github/LR.png" width=300 height=300>

$$
logit函数：\log(odds) = \log(\frac{p}{1-p})
$$

等式同时求e次方，公式如下：

$$
p = \frac{1}{1 + e^{-\log(odds)}}
$$

**因此，sigmoid从logit函数变换过来。**

### 2. 逻辑函数 sigmoid：
$$ P = \sigma(z) = \frac{1}{1+e^{-z}}$$
其中，$z = wx + b$


### 3. 目标函数（交叉熵损失函数 CrossEntropyLoss）：

假设数据之间相互独立。

$$
P(y|X) = \prod_{i=1}^N P(y_i|x_i)
$$

maximum likelihood:

$$
\text{L(w)} = \prod_{i=1}^n (\sigma(x^i))^{y^i} (1-\sigma(x^i))^{1-y^i}
$$

maximum log likelihood:

$$
\log(\text{L(w)}) = \sum_{i=1}^n \left(y^i\log(\sigma(x^i) + (1-y^i)\log(1-\sigma(x^i))\right)
$$

minimum negative log likelihood:

$$
\text{J(w)} = - \log(\text{L(w)}) = \sum_{i=1}^n \left(- y^i\log \sigma(x^i) - (1-y^i)\log(1-\sigma(x^i))\right)
$$

对目标函数**（J(w)交叉熵损失函数）**求偏导数，结果如下：

$$
\frac{\partial \text{J(w)}}{\text{w}} = \frac{1}{n} \sum_{i=1}^n \left(\sigma(x^i) - y^i)\right)x^i
$$

**通过梯度下降寻找目标函数最小值。**

## 2.1 python:

```python
from sklearn.linear_model import LogisticRegression

# 1. create dataset.
normal = np.random.multivariate_normal
# Number of samples 
nSamples = 500
s2 = 1
sgx0 = normal([0.,0.], [[s2, 0.], [0.,s2]], nSamples)
sgx1 = normal([2.,2.], [[s2, 0.], [0.,s2]], nSamples)

# setting the labels for each category
sgy0 = np.zeros((nSamples,))
sgy1 = np.ones((nSamples,))

sgx = np.concatenate((sgx0, sgx1))
sgy = np.concatenate((sgy0, sgy1))

# 2. logistic regression. 
clf = LogisticRegression(solver='lbfgs')  #clf: classifier
clf.fit(sgx, sgy)

# 3. plot.
xmin, xmax, npoints = (-6,6,51)
gridx1, gridx2 = np.meshgrid(np.linspace(xmin,xmax,npoints), np.linspace(xmin,xmax,npoints))
grid = np.c_[gridx1.flatten(), gridx2.flatten()]
prob = clf.predict_proba(grid)

plt.pcolor(gridx1,gridx2,prob[:,1].reshape(npoints,npoints))
plt.colorbar()
plt.scatter(sgx0[:,0], sgx0[:,1], alpha=0.5)
plt.scatter(sgx1[:,0], sgx1[:,1], alpha=0.5)
plt.xlabel('x1')
plt.ylabel('x2')
```

## 3. 逻辑回归与贝叶斯分类：

贝叶斯公式如下：

$$P(c|x) = \frac{P(x|c)P(c)}{P(x|c)P(c) + P(x|\bar{c})P(\bar{c})}$$

等式右边同时除以分子。

$$P(c|x) = \frac{1}{1 + \frac{P(x|\bar{c})P(\bar{c})}{P(x|c)P(c)}}$$

假设，$e^{-z} = \frac{P(x|\bar{c})P(\bar{c})}{P(x|c)P(c)} $并且$P(c)=P(\bar{c})$。

$$ e^{-z} = \frac{P(x|\bar{c})}{P(x|c)} $$

等式左右同时取log并且将负号移到右边。

$$z = \log{\frac{P(x|c)}{P(x|\bar{c})}} = \log{(odds)}$$

把z带回原式。

$$P(c|x) = \frac{1}{1 + e^{-\log{(odds)}}}$$

**结论：逻辑回归本质就是贝叶斯分类。**

## 4. 牛顿法：

### 4.1 泰勒公式：

泰勒公式：如果函数平滑且某点存在各阶导数，则可以用一个多项式来描述该点邻域的近似值。公式如下：


$$
f(x) = \sum_{i=0}^\infty f^{(n)}(x_0) \frac{1}{n!} (x-x_0)^n
$$

### 4.2 牛顿法原理：

对函数在某一点二次泰勒展开求导等于0。公式如下所示：

$$
f(x) \approx g(x) = f(x_k) + f(x_k)^\prime (x-x_k) + f(x_k)^{\prime \prime} (x-x_k)^2
$$


$$
f(x_k)^\prime + f(x_k)^{\prime \prime} (x-x_k) = 0
$$


$$
x = x_k - \frac{f^\prime (x_k)}{f^{\prime \prime} (x_k)}
$$

对于多元函数，函数极值可以通过判断海森矩阵是否为正定。
