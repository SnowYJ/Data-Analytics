## 1. 三种降维方法：

1. PCA（**线性**）: **PCA使用前提是对数据进行中心化处理**，然后求协方差C，最后对协方差进行特征值分解获得特征向量。val, vec = np.linalg.eig(C) ：**此函数返回的特征值不一定有序排列。**

2. SVD（**线性**）: **SVD不需要数据中心化处理**，直接进行奇异值分解得到的右奇异矩阵即为特征矩阵，且奇异值矩阵中的奇异值沿主对角线方向从大到小排列。u, sigma, vt = np.linalg.svd(X) ：**svd()函数返回的是v的转置。**

3. Autoencoder（**非线性**）: encoder,decoder参数共享。

## 1.1 主成份分析PCA:

### 1.1.1 step-by-step 公式推导：

选取一基底使数据在此基底上的投影的方差最大，**方差越大代表数据在此基底下的信息量越大**。

首先，**将数据中心化**，**使样本空间中每个样本的每一个特征的均值为零（$\mu = 0$），样本在基底v（单位向量）上的投影的方差为：**

$$
\sigma^2(\mathbf{X}, \mathbf{v}) = \frac{1}{n} \sum_{i=1}^n(\mathbf{v}^\top x_i - 0)^2 = \frac{1}{n}(\mathbf{v}^\top \mathbf{X})^2 = \frac{1}{n}(\mathbf{v}^\top \mathbf{X})(\mathbf{v}^\top \mathbf{X})^\top = \frac{1}{n} \mathbf{v}^\top \mathbf{X}\mathbf{X}^\top \mathbf{v}
$$

其中，
$$
\mathbf{X} = 
                \begin{bmatrix}
                        x_{1,1} & \cdots & x_{1, n}\\
                        x_{2,1} & \cdots & x_{2, n}\\
                   \vdots &  \vdots\\
                        x_{q,1} & \cdots & x_{q, n}
                \end{bmatrix},       
\mathbf{v} = 
                \begin{bmatrix}
                    v_1\\
                    v_2\\
                    \vdots\\
                    v_q\\
                \end{bmatrix}
$$
**样本个数为n，特征值个数为q。**

由于 **特征均值为0**，
$$
\frac{1}{n} \mathbf{X} \mathbf{X}^\top = \mathbf{C}
$$

所以，
$$
\sigma^2(\mathbf{X}, \mathbf{v}) = \mathbf{v}^\top \mathbf{C} \mathbf{v}
$$

通过拉格朗日乘数法求极大值。

**首先，寻找第一个特征向量。约束条件为：1.单位向量。**

$$
f(\mathbf{v}) = \mathbf{v}^\top \mathbf{C} \mathbf{v} - \lambda (\mathbf{v}^\top \mathbf{v} - 1)
$$

$$
\frac{df(\mathbf{v})}{d\mathbf{v}} = 2\mathbf{C}\mathbf{v} - 2\lambda\mathbf{v} = 0
$$

$$
\mathbf{C}\mathbf{v} = \lambda\mathbf{v}
$$

其中，v为特征向量，$\lambda$为特征值或方差（ $
\sigma^2(\mathbf{X}, \mathbf{v})=\mathbf{v}^\top\mathbf{C}\mathbf{v} = \mathbf{v}^\top \lambda \mathbf{v} = \lambda
$），C为协方差。

**接下来寻找第二个特征向量，约束条件为：1.单位向量 2.与第一个特征向量正交。**

$$
f(\mathbf{v}, \mathbf{v}_2) = \mathbf{v}_2^\top \mathbf{C} \mathbf{v}_2 - \alpha (\mathbf{v}_2^\top \mathbf{v}_2 - 1) -\beta (\mathbf{v}_2^\top \mathbf{v} - 0)
$$


**以此类推，知道寻找出k个特征向量。** https://zhuanlan.zhihu.com/p/47858230

### 总结：

step-by-step 步骤为首先根据拉格朗日乘数法寻找出第一个投影方差最大对应的特征向量。其次，增加约束条件，寻找出第二个投影方差最大对应的特征向量，以此类推。

### 1.1.2 特征分解：

**对协方差矩阵进行特征值分解可以获得特征值与特征向量。由于协方差矩阵为实对称矩阵，因此，不同的特征向量是两两正交的线性无关。**

**实对称矩阵C**可以分解为下面的形式：

$$
\mathbf{C} = \mathbf{Q} \Sigma \mathbf{Q}^\top
$$

其中，
$$
\boldsymbol{\Sigma} = \begin{bmatrix}\sigma^2_{1} & 0 & 0 & 0\\
                                     0 & \sigma^2_{2} & 0 & 0\\
                                     0 & 0 & \ddots & 0\\
                                     0 & 0 & 0 & \sigma^2_q\end{bmatrix},
\mathbf{Q} = \begin{bmatrix}
                \mathbf{v}_1 & \mathbf{v}_2 & \dots & \mathbf{v}_q
             \end{bmatrix}
$$

**Q的大小为q$\times$q，例如：数据包含20个特征，使用特征分解后的每个特征向量维度为20，一共有20个特征向量。**

### 1.1.3 特征数量远大于样本数量：

$$
\mathbf{Y} \mathbf{Y}^\top \mathbf{v} = \lambda\mathbf{v}
$$

$$
\mathbf{Y}^\top \mathbf{Y} \mathbf{u} = \sigma\mathbf{u}
$$

$$
\mathbf{Y} \mathbf{Y}^\top(\mathbf{Y} \mathbf{v}) = \sigma(\mathbf{Y}\mathbf{v}).
$$

$$
\mathbf{C}(\mathbf{Y} \mathbf{v}) = \sigma(\mathbf{Y}\mathbf{v})
$$

**协方差矩阵的特征向量等于$\mathbf{Y} \mathbf{v}$，由于，$\mathbf{Y}^\top \mathbf{Y}$为n$\times$n的矩阵，其中，n<<p。运算量大大减小节约计算空间。**

### 1.1.4 python代码：

```python
import torch
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt

trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True)
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True)

train_dataset = trainset.data / 255
x, y, z = train_dataset.shape[1:] 
dataset = train_dataset.reshape((-1, x*y*z))

# remove mean. 
mean_dataset = np.mean(dataset, axis=1).reshape((-1, 1))
mean_removed_dataset = dataset - mean_dataset

# covariance.
cov_dataset = np.cov(mean_removed_dataset, rowvar=0)
print("the shape of covariance is ", cov_dataset.shape)

# eigenvector (row vectors), eigenvalue. 
eig_vals, eig_vects = np.linalg.eig(cov_dataset)
print("the shape of eigenvector is ", eig_vects.shape)

# sort eigenvalue.
eig_index = np.argsort(eig_vals)

# top 100 eigenvalues.
eig_index_100 = eig_index[:-101:-1]
print(eig_index_100.shape)

# corresponding eigenvectors.
eig_vects_100 = eig_vects[eig_index_100]
print(eig_vects_100.shape)

DR_dataset = np.dot(mean_removed_dataset, eig_vects_100.T)
print(DR_dataset.shape)
```

```python
from sklearn.decomposition import PCA

x_train_reshape = train_dataset.reshape((-1, x*y*z))

pca_object = PCA(n_components = 300, svd_solver='randomized', whiten=True)
x_train_intmde = pca_object.fit_transform(x_train_reshape)
x_train_approx = pca_object.inverse_transform(x_train_intmde)

plt.imshow(x_train_approx[0].reshape((x, y, z)))
```

### 总结：

* PCA首先计算出协方差矩阵C，然后对其进行特征值分解获得特征向量。最后，样本空间乘以由列特征向量组成的特征矩阵（也就是SVD中的右奇异矩阵）。
* **当特征数q远大于样本数量n的情况时，将样本空间转置然后PCA降维。**

## 1.2 奇异值分解SVD:

### 1.2.1 SVD原理：

有一$n\times m$的实数矩阵A可分解为下面的形式：

$$
\mathbf{A} = \mathbf{U} \Sigma \mathbf{V}^\top
$$
其中，U，V为单位正交矩阵，其中，U为左奇异矩阵，V为右奇异矩阵。n为样本数，m为特征数。
$
\boldsymbol{\Sigma} = \begin{bmatrix}\sigma_{1} & 0 & 0 & 0\\
                                     0 & \sigma_{2} & 0 & 0\\
                                     0 & 0 & \ddots & 0\\
                                     0 & 0 & 0 & \sigma_m\end{bmatrix}，
$
对角线上的值称为奇异值。

如何求U，V以及$\Sigma$ ？

$$
\mathbf{A}\mathbf{A}^\top = \mathbf{U} \Sigma \Sigma^\top \mathbf{U}^\top
$$
$$
\mathbf{A}^\top\mathbf{A} = \mathbf{V} \Sigma \Sigma^\top \mathbf{V}^\top 
$$

$\mathbf{A}\mathbf{A}^\top, \mathbf{A}^\top\mathbf{A}$均为实对称矩阵。对其分别进行特征值分解，可以获得U，V，$\sigma$。

**注意：左奇异矩阵对数据个数（行）进行降维，右奇异矩阵对特征进行降维。**

### 1.2.2 python代码:

```python
# using dataset rather than mean_removed_dataset.
# 1.
eig_u_val, eig_u_vec = np.linalg.eig(dataset[:1000]@dataset[:1000].T)
eig_v_val, eig_v_vec = np.linalg.eig(dataset[:1000].T@dataset[:1000])
sig = np.sqrt(eig_u_val)

# 2.
u, sigma, vt = np.linalg.svd(dataset[:1000])# u: column vector vt: row vector.

print("the shape of u is ", u.shape)
print("the shape of sigma is", sigma.shape)
print("the shape of vt is ", vt.shape)

# top k.
k = 100
new_dataset = (u[:, 0:k]@np.diag(sigma[:k])@vt[0:k, :]).reshape((-1, x, y, z))
# print(new_dataset.shape) notice: shape of new_dataset will not be changed.

plt.figure()
plt.title("original image")
plt.imshow(train_dataset[0])
plt.figure()
plt.title("reconstructed image")
plt.imshow(new_dataset[0])
```

### 总结：

* SVD不需要对数据进行中心化处理。
* 奇异值在奇异值矩阵中是从大到小排列的，并且奇异值减小特别快。通常，前10%或1%的奇异值占全部奇异值之和的99%。
* sklearn中PCA函数使用SVD右奇异矩阵降维。
* U与V的区别在于，U对行数进行压缩，V对列（特征数）进行压缩。

## 1.3 自编码器 Autoencoder:

<img src="ml_github/autoencoder.png" width="500" height="500">

目标函数：
$$
\left(\text{input}-\text{output}\right)^2
$$

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
from torchvision import datasets, transforms

# MNIST: picture (size 28 by 28).
mnist_data = datasets.MNIST('data', train=True, download=True, transform=transforms.ToTensor())
print(len(mnist_data))
mnist_data = list(mnist_data)[:4096]
print(len(mnist_data))
```

```python
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            # 1 input channel 2. number of kernal 3. size of kernal 3X3
            nn.Conv2d(1, 16, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 64, 7) # why this layer dont have activation function ??
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 7),
            nn.ReLU(),
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(16, 1, 3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid()  #to range [0, 1]
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
myAE=Autoencoder()
print(myAE)
```

### 参数分析：

<img src="ml_github/CNN.png" width="800" height="800">

* input: 32 by 32.
* C1: kernal: 6 $\times$ 5 by 5, output: 6 $\times$ 28 by 28. 参数：6 $\times$ (5 $\times$ 5 + 1) = 156
* S2: 
* C3: kernal: 6 $\times$ 5 by 5, output: 16 $\times$ 10 by 10. 例如：16个特征图中，前6个由上一层中前3个得到。中间6个由上一层中前4个得到。最后4个由上一层中全部的特征图得到。因此，参数个数为：6$\times$(5$\times$5$\times$3+1)+6$\times$(5$\times$5$\times$4+1)+ 4$\times$(5$\times$5$\times$6+1)
* S4:
* C5: kernal: 16 $\times$ 5 by 5, output: 120 $\times$ 1 by 1. 参数：120$\times$(5$\times$5$\times$16+1)=48120
* F6: output: 84. 参数：120$\times$84+84 = 10164
* output: 产生10类。

```python
params = list(myAE.parameters())
"""
1. input 28 by 28; kernal size: 3 by 3 by 1; 16 kernals; 3*3*16 weight; outputs: 16 * 14 by 14. output=(input-kernal)/stride + 1
"""
print(params[0].shape) # [16, 1, 3, 3] 
"""
2. bias: 16. 
"""
print(params[1].shape) # [16] 
"""
3. input 14 by 14; kernal size: 3 by 3 by 16; 32 kernals; 32*16*3*3 weight; output: 32 * 7 by 7 by 16.
"""
print(params[2].shape) # [32, 16, 3, 3]
"""
4. bias: 32.
"""
print(params[3].shape)
"""
5. input: 7 by 7 by 32; kernal size: 7 by 7 by 32; 64 kernals; 7*7*32*64 weight.
"""
print(params[4].shape)

print(params[5].shape) # code
print(params[6].shape)
```

```python
#Hyperparameters for training
batch_size=64
learning_rate=1e-3
max_epochs = 20

#Set the random seed for reproducibility 
torch.manual_seed(509) 
#Choose mean square error loss
criterion = nn.MSELoss() 
#Choose the Adam optimiser
optimizer = torch.optim.Adam(myAE.parameters(), lr=learning_rate, weight_decay=1e-5)
#Specify how the data will be loaded in batches (with random shffling)
train_loader = torch.utils.data.DataLoader(mnist_data, batch_size=batch_size, shuffle=True) # mini-batch
#Storage
outputs = []

#Start training
for epoch in range(max_epochs):
    for data in train_loader:
        img, label = data # img: input
        optimizer.zero_grad() # zero parameter gradient
        recon = myAE(img) 
        loss = criterion(recon, img)
        loss.backward() # calculate gradient by backward proporgation.
        optimizer.step() # update parameters.           
    if (epoch % 3) == 0:
        print('Epoch:{}, Loss:{:.4f}'.format(epoch+1, float(loss)))
    outputs.append((epoch, img, recon),)
```

```python
numImgs=12
for k in range(0, max_epochs, 9):
    plt.figure(figsize=(numImgs, 2))
    imgs = outputs[k][1].detach().numpy() # 截断反向传播梯度流
    recon = outputs[k][2].detach().numpy()
    for i, item in enumerate(imgs):
        if i >= numImgs: break
        plt.subplot(2, numImgs, i+1)
        plt.imshow(item[0])
        
    for i, item in enumerate(recon):
        if i >= numImgs: break
        plt.subplot(2, numImgs, numImgs+i+1)
        plt.imshow(item[0])
```

```python
"""t-SNE对手写数字进行可视化"""
from time import time
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.manifold import TSNE


def get_data():
    digits = datasets.load_digits(n_class=6)
    data = digits.data
    label = digits.target
    n_samples, n_features = data.shape
    return data, label, n_samples, n_features


def plot_embedding(data, label, title):
    x_min, x_max = np.min(data, 0), np.max(data, 0)
    data = (data - x_min) / (x_max - x_min)

    fig = plt.figure()
    ax = plt.subplot(111)
    for i in range(data.shape[0]):
        plt.text(data[i, 0], data[i, 1], str(label[i]),
                 color=plt.cm.Set1(label[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})
    plt.xticks([])
    plt.yticks([])
    plt.title(title)
    return fig


def main():
    data, label, n_samples, n_features = get_data()
    print('Computing t-SNE embedding')
    tsne = TSNE(n_components=2, init='pca', random_state=0)
    t0 = time()
    result = tsne.fit_transform(data)
    fig = plot_embedding(result, label,
                         't-SNE embedding of the digits (time %.2fs)'
                         % (time() - t0))
    plt.show(fig)


if __name__ == '__main__':
    main()
```

```python
from time import time

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter

from sklearn import manifold, datasets

# # Next line to silence pyflakes. This import is needed.
# Axes3D

n_points = 1000
X, color = datasets.samples_generator.make_s_curve(n_points, random_state=0) # X: 1000 by 3; color(label)
n_neighbors = 10
n_components = 2 # expected dimensionality.

fig = plt.figure(figsize=(8, 8))
plt.suptitle("Manifold Learning with %i points, %i neighbors" % (1000, n_neighbors), fontsize=14)

'''绘制S曲线的3D图像'''
ax = fig.add_subplot(211, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
ax.view_init(4, -72)  # 初始化视角

'''t-SNE'''
t0 = time()
tsne = manifold.TSNE(n_components=n_components, init='pca', random_state=0)
Y = tsne.fit_transform(X) 
t1 = time()
print("t-SNE: %.2g sec" % (t1 - t0))
ax = fig.add_subplot(2, 1, 2)
plt.scatter(Y[:, 0], Y[:, 1], c=color, cmap=plt.cm.Spectral)
plt.title("t-SNE (%.2g sec)" % (t1 - t0))
ax.xaxis.set_major_formatter(NullFormatter())  # 设置标签显示格式为空
ax.yaxis.set_major_formatter(NullFormatter())

plt.show()
```

## 2. k-means:

目标函数（k个类别）：
$$
E=\sum_{j=1}^k \sum_{i\ \text{allocated to}\ j}  \left(\mathbf{y}_{i, :} - \boldsymbol{\mu}_{j, :}\right)^\top\left(\mathbf{y}_{i, :} - \boldsymbol{\mu}_{j, :}\right)
$$

算法步骤：
1. 随机选择k个样本作为k个分类的中心点。
2. 计算每个样本距离中心点的欧式平方距离。
3. 根据最短距离，将样本分配给k个类。
4. 重新计算中心点。
5. 重复上述步骤，直至样本中心点不在发生改变。

```python
from sklearn.cluster import KMeans
# dataset.
mnist_data = datasets.MNIST('data', train=True, download=True)
x_train_mnist = mnist_data.data.numpy()
y_train_mnist = mnist_data.targets.numpy()

# subset.
x_train_mnist_2_4 = x_train_mnist[(y_train_mnist == 2) | (y_train_mnist == 4)]
x_train_mnist_2_4_reshape = x_train_mnist_2_4.reshape((-1, x_train_mnist.shape[1]*x_train_mnist.shape[2]))

# pca.
pca_object = PCA(n_components = 2, svd_solver='randomized', whiten=True)
# dataset after pca.
x_train_mnist_2d = pca_object.fit_transform(x_train_mnist_2_4_reshape)
# k-means
kmeans = KMeans(n_clusters=2, random_state=130).fit(x_train_mnist_2d)
# plot.
plt.scatter(x_train_mnist_2d[:, 0][kmeans.labels_ == 1], x_train_mnist_2d[:, 1][kmeans.labels_ == 1], c = 'r', label = '2')#
plt.scatter(x_train_mnist_2d[:, 0][kmeans.labels_ == 0], x_train_mnist_2d[:, 1][kmeans.labels_ == 0], c = 'b', label = '4')#
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1])
plt.title("K means fit")
plt.legend()
```

### 3.3 卷积神经网络中重要概念：

* 局部连接层 Locally-connected layers：减小参数数量。
<img src="ml_github/local_connected.png" width="700" height="500">


* 权重共享：每一个卷积核的参数不会发生随着扫过图片位置的变化而改变。因此，每一个卷积核可以提取处图片中某一特征。
<img src="ml_github/weight_share.png" width="200" height="300">

* 三维卷积核：卷积核的深度等于输入层的输入。例如：输入图片为RGB图片，提取6个特征图。因此，第一个卷积层需要6个卷积核，其中每个卷积核的深度为3。**注意：下一个卷积层的卷积核深度不一定为6。**
<img src="ml_github/3d.png" width="300" height="300">

* 步长 stride：卷积核每次滑过图片的长度，下图步长为2。
<img src="ml_github/stride.png" width="500" height="500">


* 零填充 zero-padding：由于在卷积作用下，下一层的输入尺寸会有所减小。随着层数的增加，输入会越来越小，因此，零填充会保证输入大小维持在某一范围内。此外，零填充可以保留图片的边缘特征的信息。
<img src="ml_github/zero_padding.png" width="300" height="300">


* 反卷积 transpose convolution：反卷积依旧是卷积运算，只不过需要对输入进行填充。
