## 参数估计：

* 参数估计：

假设某一数据集里有1000个样本，每个样本包含100个特征，记为：$f_{j=1,\dots,n}$。假设每个特征都服从某一分布，也就是说，此数据集里有100个特征，每个特征服从某一特定的分布，参数估计目标为寻找出这些分布的参数，例如：正态分布的均值方差。使用极大似然估计可以寻找出这些特征分布的参数。

* 极大似然估计 Maximum Likelihood Extimation(MLE)：

$$
L(\theta_{c, f_j}) = \prod_{i=1}^{N=1000} \text{P}(x_{i, j}|\theta_{c, f_j})
$$

目标：寻找出最大的似然函数对应的参数（$\theta$）。

**对目标函数（L）的参数求导，令结果等于0。**

## 2. 朴素贝叶斯：MAP

$$
P(C~|~{\rm features}) = \frac{P({\rm features}~|~C)P(C)}{P({\rm features})}
$$
假设特征之间相互独立且服从同一分布。
$$
P(C~|~{\rm features}) = P(C) \prod_{i=1}^n P({\rm feature}_i~|~C)
$$

### 2.1 高斯朴素贝叶斯：

当特征值为离散值时，可以通过统计个数来计算概率。当特征值为连续值时，可以用某一值表示某一范围。例如：身高小于160cm用0表示，身高大于160小于180用1表示，大于180用2表示。但是，这种方式计算出的概率并不准确。因此，可以假设连续变量服从高斯分布。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.naive_bayes import GaussianNB

# generate training set.
X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5) # generate 100 sample from 2 categories.
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')

model = GaussianNB()
model.fit(X, y)

# generate new sample.
rng = np.random.RandomState(0)
Xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)  #the numbers are to control the x,y limits
ynew = model.predict(Xnew)

# plot prediction boundary.
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
lim = plt.axis()
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew, s=20, cmap='RdBu', alpha=0.1)
```

### 2.2 多项式朴素贝叶斯：

**多项式朴素贝叶斯多用于文本分类。注意，由于特征值为离散值，因此需要考虑拉普拉斯平滑laplacian smoothing可以有效避免概率为零的情况。**

多项式与伯努利分类器的计算粒度不一样，多项式模型以单词为粒度，伯努利模型以文件为粒度，因此二者的先验概率和类条件概率的计算方法都不同。计算后验概率时，对于一个文档d，多项式模型中，只有在d中出现过的单词，才会参与后验概率计算，伯努利模型中，没有在d中出现，但是在全局单词表中出现的单词，也会参与计算，不过是作为“反方”参与的。（在Text Processing中会被介绍）

https://blog.csdn.net/qq_27009517/article/details/80044431

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix
import seaborn as sns

# dataset.
data = fetch_20newsgroups('./data/20news')
data.target_names

categories = ['talk.religion.misc', 'soc.religion.christian', 'sci.space', 'comp.graphics']
train = fetch_20newsgroups(subset='train', categories=categories)
test = fetch_20newsgroups(subset='test', categories=categories)
train.target_names

# tfidf.
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(train.data, train.target)
labels = model.predict(test.data)

mat = confusion_matrix(test.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, xticklabels=train.target_names, yticklabels=train.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label')
```

## 3. When to Use Naive Bayes

Because naive Bayes classifiers make stringent assumptions about data, they will generally not perform as well as a more complicated model when sufficient data is available.
That said, they have several advantages:

- They are extremely fast for both training and prediction
- They provide straightforward probabilistic prediction
- They are often very easily interpretable
- They have very few (if any) tunable parameters

These advantages mean a naive Bayes classifier is often a good choice as an initial baseline classification.
If it performs suitably, then congratulations: you have a very fast, very interpretable classifier for your problem.
If it does not perform well, then you can begin exploring more sophisticated models, with some baseline knowledge of how well they should perform.

Naive Bayes classifiers tend to perform especially well in one of the following situations:

- When the naive assumptions actually match the data (very rare in practice)
- For very well-separated categories, when model complexity is less important
- For very high-dimensional data, when model complexity is less important

The last two points seem distinct, but they actually are related: as the dimension of a dataset grows, it is much less likely for any two points to be found close together (after all, they must be close in *every single dimension* to be close overall).
This means that clusters in high dimensions tend to be more separated, on average, than clusters in low dimensions, assuming the new dimensions actually add information.
For this reason, simplistic classifiers like naive Bayes tend to work as well or better than more complicated classifiers as the dimensionality grows: once you have enough data, even a simple model can be very powerful.
